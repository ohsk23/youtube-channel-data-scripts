```python
import json
import os
import math

def calculate_api_calls(comments_count):
    # 한 번에 100개씩 가져올 수 있으므로, 올림 처리하여 필요한 API 호출 횟수 계산
    return math.ceil(comments_count / 100)

# data/comments 디렉토리의 모든 .json 파일 읽기
comments_dir = './data/0-100'
total_comments = 0
total_api_calls = 0

for filename in os.listdir(comments_dir):
    if filename.endswith('.json'):
        file_path = os.path.join(comments_dir, filename)
        with open(file_path, 'r', encoding='utf-8') as f:
            comments = json.load(f)
            comments_count = len(comments)
            api_calls = calculate_api_calls(comments_count)
            
            total_comments += comments_count
            total_api_calls += api_calls
            
            print(f"{filename}: {comments_count} comments (API calls: {api_calls})")

print(f"\nTotal comments: {total_comments}")
print(f"Total API calls: {total_api_calls}")
```
